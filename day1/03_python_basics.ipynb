## 1. Variable Definitions

Variables store data values. Python is dynamically typed, so you don't need to declare the type explicitly.

```python
# Integer
age = 25
print(f"Age: {age}, Type: {type(age)}")

# Float (decimal number)
price = 19.99
print(f"Price: {price}, Type: {type(price)}")

# String (text)
name = "Alice"
print(f"Name: {name}, Type: {type(name)}")

# Boolean (True/False)
is_student = True
print(f"Is student: {is_student}, Type: {type(is_student)}")
```

```python
# Variables can be reassigned and even change types
x = 10
print(f"x = {x}")

x = "now I'm a string"
print(f"x = {x}")

# Multiple assignment
a, b, c = 1, 2, 3
print(f"a={a}, b={b}, c={c}")
```

## 2. Lists

Lists are ordered, mutable collections that can hold items of different types.

```python
# Creating lists
fruits = ["apple", "banana", "cherry"]
numbers = [1, 2, 3, 4, 5]
mixed = [1, "hello", 3.14, True]

print(f"Fruits: {fruits}")
print(f"Numbers: {numbers}")
print(f"Mixed types: {mixed}")
```

```python
# Accessing elements (0-indexed)
print(f"First fruit: {fruits[0]}")
print(f"Last fruit: {fruits[-1]}")

# Slicing
print(f"First two fruits: {fruits[0:2]}")
print(f"Middle numbers: {numbers[1:4]}")
```

```python
# Modifying lists
fruits.append("orange")       # Add to end
print(f"After append: {fruits}")

fruits.insert(1, "mango")     # Insert at position
print(f"After insert: {fruits}")

fruits.remove("banana")       # Remove by value
print(f"After remove: {fruits}")

popped = fruits.pop()         # Remove and return last item
print(f"Popped: {popped}, List: {fruits}")

print(f"Length of list: {len(fruits)}")
```

## 3. Loops

Loops let you execute code repeatedly. Python has `for` and `while` loops.

```python
# For loop - iterate over a sequence
colors = ["red", "green", "blue"]

for color in colors:
    print(f"Color: {color}")
```

```python
# For loop with range()
print("Counting 0 to 4:")
for i in range(5):
    print(i)

print("\nCounting 2 to 6:")
for i in range(2, 7):
    print(i)

print("\nCounting by 2s:")
for i in range(0, 10, 2):
    print(i)
```

```python
# enumerate() gives you index and value
animals = ["cat", "dog", "bird"]

for index, animal in enumerate(animals):
    print(f"{index}: {animal}")
```

```python
# While loop - repeat while condition is True
count = 0

while count < 5:
    print(f"Count is {count}")
    count += 1  # Same as count = count + 1

print("Done!")
```

```python
# break and continue
print("Using break:")
for i in range(10):
    if i == 5:
        break  # Exit the loop
    print(i)

print("\nUsing continue:")
for i in range(5):
    if i == 2:
        continue  # Skip this iteration
    print(i)
```

## 4. Logic Statements (Conditionals)

Control flow with `if`, `elif`, and `else` statements.

```python
# Basic if statement
temperature = 75

if temperature > 80:
    print("It's hot!")
elif temperature > 60:
    print("It's nice outside.")
else:
    print("It's cold!")
```

```python
# Comparison operators
x = 10
y = 5

print(f"x == y: {x == y}")   # Equal to
print(f"x != y: {x != y}")   # Not equal to
print(f"x > y:  {x > y}")    # Greater than
print(f"x < y:  {x < y}")    # Less than
print(f"x >= y: {x >= y}")   # Greater than or equal
print(f"x <= y: {x <= y}")   # Less than or equal
```

```python
# Logical operators: and, or, not
age = 25
has_license = True

if age >= 18 and has_license:
    print("You can drive.")

is_weekend = False
is_holiday = True

if is_weekend or is_holiday:
    print("No work today!")

is_raining = False

if not is_raining:
    print("No umbrella needed.")
```

```python
# Checking membership with 'in'
fruits = ["apple", "banana", "cherry"]

if "banana" in fruits:
    print("We have bananas!")

if "mango" not in fruits:
    print("No mangos available.")
```

## 5. Functions

Functions are reusable blocks of code that perform specific tasks.

```python
# Basic function definition
def greet():
    print("Hello, World!")

# Calling the function
greet()
```

```python
# Function with parameters
def greet_person(name):
    print(f"Hello, {name}!")

greet_person("Alice")
greet_person("Bob")
```

```python
# Function with return value
def add(a, b):
    return a + b

result = add(3, 5)
print(f"3 + 5 = {result}")

# You can use the result directly
print(f"10 + 20 = {add(10, 20)}")
```

```python
# Default parameter values
def greet_with_message(name, message="Welcome!"):
    print(f"{message} {name}")

greet_with_message("Alice")                    # Uses default
greet_with_message("Bob", "Good morning,")     # Custom message
```

```python
# Function returning multiple values
def get_stats(numbers):
    total = sum(numbers)
    count = len(numbers)
    average = total / count
    return total, count, average

data = [10, 20, 30, 40, 50]
total, count, avg = get_stats(data)

print(f"Total: {total}")
print(f"Count: {count}")
print(f"Average: {avg}")
```

```python
# Putting it all together - a practical example
def analyze_grades(grades):
    """Analyze a list of grades and return summary statistics."""
    if not grades:  # Check if list is empty
        return "No grades to analyze"
    
    total = 0
    passing = 0
    
    for grade in grades:
        total += grade
        if grade >= 60:
            passing += 1
    
    average = total / len(grades)
    pass_rate = (passing / len(grades)) * 100
    
    return {
        "average": average,
        "highest": max(grades),
        "lowest": min(grades),
        "pass_rate": pass_rate
    }

# Test the function
student_grades = [85, 92, 78, 55, 90, 67, 73, 88]
results = analyze_grades(student_grades)

print("Grade Analysis:")
print(f"  Average: {results['average']:.1f}")
print(f"  Highest: {results['highest']}")
print(f"  Lowest: {results['lowest']}")
print(f"  Pass Rate: {results['pass_rate']:.1f}%")
```
